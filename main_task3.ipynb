## Data Visualisation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns #For plotting advance charts
from datetime import datetime
```

```python
data = pd.read_csv('householdtask3.csv')
```

```python
display(data.head(10))
```

### Scatter plot with year against own

```python
plt.scatter(data['year'],data['own'])

#Adding title to the plot
plt.title("Scatter plot")

#Setting the x & y labels
plt.xlabel('year')
plt.ylabel('own')

#adding the legends
plt.show()
```

### Line chart with year against own

```python
plt.plot(data['year'])
plt.plot(data['own'])

#Adding title to the plot
plt.title("Line plot")

#Setting the x & y labels
plt.xlabel('year')
plt.ylabel('own')

#adding the legends
plt.show()
```

### Bar Chart or Bar plot

```python
plt.bar(data['year'],data['own'])

#Adding title to the plot
plt.title("Bar plot")

#Setting the x & y labels
plt.xlabel('year')
plt.ylabel('own')

#adding the legends
plt.show()
```

### Histogram

```python
plt.hist(data['income'])

plt.title("Histogram")

plt.show()
```

### Pie Chart

```python
# Creating plot
fig = plt.figure(figsize=(10, 7))
plt.pie(data['income'], labels=data['year'])

plt.show()
```

### Heat Map

```python
plt.figure(figsize=(10,5))
sns.heatmap(data.corr())
```

### Box Plot 

```python
figure = plt.figure(figsize =(10, 7))  
ax = figure.add_axes([0, 0, 1, 1])  
bp = ax.boxplot(data)  
plt.show()  
```
